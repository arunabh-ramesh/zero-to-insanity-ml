```python
#Create a vector
v = [3, 2]
print(v)
```

```python
#Multiply the vector by a scalar value (lambda)
result = [v[0] * 2, v[1] * 2]
print(result)
```

```python
#Add 2 vectors together
w = [1, 4]
result = [v[0] + w[0], v[1] + w[1]]
print(result)
```

```python
#Dot product of two vectors
result = [v[0] * w[0], v[1] * w[1]]
print(result)
```

```python
#Matrix multiplication with a single vector
v = [3, 2]
matrix = [[2, -1], [1, 1]]
result = [v[0]*matrix[0][0] + v[1]*matrix[0][1], v[0]*matrix[1][0] + v[1]*matrix[1][1]]
print(result)
```

```python
#Matrix multiplication function
def matrixMulti(v, m):
    result = [];
    for i in m:
        total = 0
        for j in range(len(v)):
            total += v[j] * i[j]
        result.append(total)
    return result
print(matrixMulti(v, matrix))
```
